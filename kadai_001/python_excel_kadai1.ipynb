```python
import openpyxl
from datetime import date

wb = openpyxl.Workbook()
ws = wb.active

ws['B2'] = "請求書"
ws['B4'] = '株式会社ABC'
ws['B5'] = '〒101-0022 東京都千代田区神田練塀町300'
ws['B6'] = 'TEL:03-1234-5678 FAX:03-1234-5678'
ws['B7'] = '担当者名:鈴木一郎 様'

ws['F4'] = 'No.'
ws['G4'] = '0001'
ws['F5'] = '日付'
ws['G5'] = date.today().strftime('%Y/%m/%d') # 今日の日付を取得

ws['B10'] = '商品名'
ws['C10'] = '数量'
ws['D10'] = '単価'
ws['E10'] = '金額'

ws['B11'] = '商品A'
ws['C11'] = 2
ws['D11'] = 10000
ws['E11'] = 20000 # C11 * D11

ws['B12'] = '商品B'
ws['C12'] = 1
ws['D12'] = 15000
ws['E12'] = 15000 # C12 * D12

ws['E13'] = ws['E11'].value + ws['E12'].value

ws['B15'] = '税込合計'
ws['E15'] = ws['E13'].value # 合計金額
ws['B16'] = '消費税'
ws['E16'] = ws['E13'].value * 0.1 # 消費税額
ws['E17'] = ws['E15'].value + ws['E16'].value # 税込合計

file_name = '請求書_20260405.xlsx'

wb.save(file_name)
```
