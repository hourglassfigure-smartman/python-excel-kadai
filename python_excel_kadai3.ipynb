```python
import pandas as pd
from openpyxl.styles import PatternFill

# 1. Excelファイルの読み込みとデータフレームの作成
df_2022 = pd.read_excel('/content/2022_年間売上表.xlsx', sheet_name='Sheet1', engine='openpyxl')
df_2023 = pd.read_excel('/content/2023_年間売上表.xlsx', sheet_name='Sheet1', engine='openpyxl')

# 2. データフレームの連結
df_combined = pd.concat([df_2022, df_2023], ignore_index=True)

# 3. 「商品」と「売上年」ごとにグループ化し、「金額（千円）」を集計
df_grouped = df_combined.groupby(['商品', '売上年'])['金額（千円）'].sum().reset_index()

# 4. 新規Excelファイル「売上集計表.xlsx」に書き込み、ヘッダーをスタイリング
output_file = '売上集計表.xlsx'

with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
    df_grouped.to_excel(writer, sheet_name='Sheet1', index=False)

    workbook = writer.book
    sheet = writer.sheets['Sheet1']

    header_fill = PatternFill(start_color='F2F2F2', end_color='F2F2F2', fill_type='solid')

    for cell in sheet[1]:
        cell.fill = header_fill

print(f'データは「{output_file}」に書き込まれ、ヘッダーがスタイル設定されました。')
print('\nすべての処理が完了しました。')
```
